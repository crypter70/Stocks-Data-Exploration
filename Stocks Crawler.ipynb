# Import Library

```python
import yfinance as yf
import pandas as pd
from datetime import datetime as dt
import datetime
```

# Price Crawling

```python
now = datetime.datetime.today().strftime('%d_%m_%Y')
end = dt.now()

# start = dt(end.year - 2, end.month, end.day)
start = '2020-01-01'

stock_code = ['BBCA.JK', 'BBNI.JK', 'BUMI.JK', 'INDY.JK', 'ADRO.JK']
    
for stock in stock_code:
    data = yf.download(stock, start=start, end=end)
    data = data[['Open', 'High', 'Low', 'Adj Close', 'Volume']]
    title = stock.replace('JK', '').replace('.', '')
    df = pd.DataFrame(data)
    df.to_csv(f'data\\{title}_{now}.csv')
```

```python
df_cf = unvr.cashflow
df_cf = df_cf.T
df_cf = df_cf[['Net Income', 'Investments']]

date = ['2021', '2020', '2019', '2018']
df_cf['Date'] = date
df_cf = df_cf.set_index("Date")
df_cf
```

```python
df_bs = unvr.balance_sheet
df_bs = df_bs.T
df_bs = df_bs[['Total Liab', 'Total Assets', 'Cash']]

date = ['2021', '2020', '2019', '2018']

df_bs['Date'] = date
df_bs = df_bs.set_index("Date")

df_bs
```

```python
income = unvr.earnings
df_inc = pd.DataFrame(income)

date = ['2018', '2019', '2020', '2021']

df_inc['Date'] = date
df_inc = df_inc.set_index("Date")
df_inc = df_inc.sort_values(by=['Date'], ascending=False)

df_inc.head()
```

```python
df_bbca = (pd.merge(df_cf, df_bs, left_index=True, right_index=True))
df_bbca = (pd.merge(df_bbca, df_inc, left_index=True, right_index=True))
df_bbca
```
